Q1. What is Machine Learning? Explain the main types of Machine Learning with examples.

Machine Learning (ML) is a subset of Artificial Intelligence (AI) that enables computers to learn from data and make decisions or predictions without being explicitly programmed. Instead of writing rigid, hard-coded rules, we feed data into algorithms, allowing them to uncover hidden patterns and improve their performance over time.

There are three primary types of Machine Learning:
1. Supervised LearningConcept: The model is trained on labeled data, meaning every training input comes with its corresponding correct output. The goal is to learn a mapping function from inputs ($X$) to outputs ($Y$).Example: Predicting whether an email is "Spam" or "Not Spam", or predicting house prices based on features like area and location.
2. Unsupervised LearningConcept: The model works with unlabeled data. There is no target output provided; the algorithm tries to find inherent structures, groupings, or patterns within the data on its own.Example: Customer segmentation for marketing (grouping buyers by similar spending habits), or anomaly detection for fraud.
3. Reinforcement Learning (RL)Concept: The model (called an agent) learns to make decisions by interacting with an environment. It performs actions and receives feedback in the form of rewards (for correct actions) or penalties (for wrong actions) to learn the optimal strategy.Example: Training an AI to play chess, or teaching self-driving cars how to navigate traffic.

Q3. What is Linear Regression? Write the Linear Regression equation and explain intercept and coefficient

Linear Regression is a supervised machine learning algorithm used to model the relationship between a dependent variable (target) and one or more independent variables (features) by fitting a linear equation to the observed data. The goal is to find the "line of best fit" that minimizes the distance between the actual data points and the predicted values on the line.

The Linear Regression Equation

For Simple Linear Regression (one independent variable), the equation is written as:
$$y = \beta_0 + \beta_1 x + \epsilon$$For Multiple Linear Regression (multiple independent variables), the equation expands to:
$$y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n + \epsilon$$

Where:
- $y$ is the Dependent Variable / Target (the value we want to predict).
- $x_1, x_2, \dots, x_n$ are the Independent Variables / Features (the inputs used for prediction).
- $\beta_0$ is the Intercept.
- $\beta_1, \beta_2, \dots, \beta_n$ are the Coefficients (or weights) of the features.
- $\epsilon$ is the Error term (Residual), representing the variance in $y$ that cannot be explained by the linear relationship with $x$.

1. Intercept ($\beta_0$)Definition:

- The intercept is the expected or predicted value of $y$ when all independent variables ($x$) are equal to zero.
- Graphical Meaning: It is the exact point where the regression line crosses the y-axis.
- Example: In a house price prediction model, if $x$ represents the number of rooms, an intercept of $\$50,000$ means a base property value of $50,000 before accounting for any rooms.

2. Coefficient ($\beta_1, \beta_2$, etc.)Definition:

 - A coefficient represents the change in the dependent variable ($y$) for every one-unit increase in its corresponding independent variable ($x$), assuming all other variables remain constant.
 - Graphical Meaning: It represents the slope of the regression line (how steep or flat it is).
 - Significance: * A positive coefficient means a direct relationship: as $x$ increases, $y$ increases.
 - A negative coefficient means an inverse relationship: as $x$ increases, $y$ decreases.
 - Example: If the coefficient for the feature "number of rooms" is +25,000, it means that adding one additional room adds an estimated $25,000 to the house price.

Q4. Explain the following evaluation metrics used in Linear Regression:
- SSE
- MSE
- RMSE
- R² Score
- Adjusted R²

1. $SSE$ (Sum of Squared Errors)

$What\ it\ is:$ The sum of the squared differences between the actual target values and the predicted values.

Formula:$$\text{SSE} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

$Interpretation:$ It gives a raw, un-averaged measure of the total error in the model. Squaring the errors ensures that positive and negative errors don't cancel each other out and penalizes larger mistakes more heavily. Lower is better.

2. $MSE$ (Mean Squared Error)

$What\ it\ is:$ The average of the squared errors across all data points. It is simply the SSE divided by the total number of observations ($n$).

Formula:$$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

$Interpretation:$ MSE is the standard loss function used to train linear regression models. Because it squares the errors, it is highly sensitive to outliers. A single massive prediction error will blow up the MSE. Lower is better.

3. $RMSE$ (Root Mean Squared Error)

$What\ it\ is:$ The square root of the Mean Squared Error.

Formula:$$\text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$

$Interpretation:$ While MSE is great for training, its units are squared (e.g., if you are predicting house prices in dollars, MSE is in "dollars squared," which makes no sense to interpret). RMSE brings the metric back to the original unit of the target variable, making it highly intuitive. Lower is better.

4. $R^2\ Score$ (Coefficient of Determination)

$What\ it\ is:$ A metric that represents the proportion of variance in the dependent variable that is predictable from the independent variables. It compares your model against a baseline model that just predicts the mean value of $y$.

Formula:$$R^2 = 1 - \frac{\text{SSE}}{\text{SST}}$$(where $\text{SST} = \sum (y_i - \bar{y})^2$ is the Total Sum of Squares)

$Interpretation:$ It generally ranges from 0 to 1 (though it can be negative if the model is worse than a simple horizontal line at the mean). An $R^2$ of 0.85 means your model explains 85% of the variance in the target data. Higher is better.

5. $Adjusted\ R^2$

$What\ it\ is:$ A modified version of $R^2$ that accounts for the number of predictors (features) in the model.

Formula:$$\text{Adjusted } R^2 = 1 - \left[ \frac{(1 - R^2)(n - 1)}{n - p - 1} \right]$$(where $n$ is the number of data points and $p$ is the number of independent variables)

$Interpretation:$ Standard $R^2$ has a major flaw: it will always increase or stay the same when you add new features, even if those features are pure garbage (like adding "favorite color" to predict house prices). Adjusted $R^2$ penalizes you for adding irrelevant features. It will only increase if the new feature improves the model more than what would be expected by pure chance. Higher is better.

Q5. Explain the Machine Learning workflow steps from data collection to model evaluation.

1. Data Collection
- This is the initial step where you gather raw data from various sources such as databases, CSV files, APIs, web scraping, or IoT devices.
- Context: For this project, the data comes from the structured Boston Housing dataset.
2. Data Cleaning & Preprocessing
- Raw data is rarely perfect. This step handles real-world anomalies to get the data model-ready.
- Key operations include handling missing values, identifying and managing outliers, resolving duplicate records, and converting data types where necessary.
3. Exploratory Data Analysis (EDA)
- This phase focuses on discovering patterns, spotting anomalies, and checking assumptions using summary statistics and graphical visualizations.
- Analysts use tools like correlation heatmaps and pairplots to understand structural dependencies and identify which features share a strong relationship with the target variable.
4. Feature Engineering & Selection
- This involves preparing your variables for the model.
- Tasks include normalizing or scaling features (ensuring variables like "number of rooms" and "property tax" are comparable), encoding categorical variables, and selecting the most predictive features while dropping redundant ones.
5. Data Splitting
- To ensure the model can generalize to unseen data, the clean dataset is divided into subsets.
- The standard practice is an 80/20 train-test split , where 80% is used to train the model's parameters and 20% is held out to evaluate true performance.
6. Model Training & EvaluationTraining
- The selected algorithm (e.g., Linear Regression) processes the training subset to find the mathematical weights or line of best fit.
- Evaluation: Predictions are generated using the remaining test set. The model's accuracy is quantified using metrics such as MAE, MSE, RMSE, and $R^2$ to decide if it is ready for deployment.

Q8. Select independent variables (features) and dependent variable (target)

Target variable: medv (house price)

```python
# Target variable: medv (house price)
y = df['medv']

# Independent variables: everything else except the target
X = df.drop(columns=['medv'])

# Verify the selection
print("Independent Variables (Features):", list(X.columns))
print("Dependent Variable (Target):", y.name)
```

Q9. Split the dataset into training and testing sets (80% training, 20% testing).

```python
from sklearn.model_selection import train_test_split

# Split the dataset into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Verify the split dimensions
print(f"Training features size: {X_train.shape}")
print(f"Testing features size: {X_test.shape}")
```

Q10. Train a Linear Regression model and perform the following:
Make predictions
- Calculate MAE
- Calculate MSE
- Calculate RMSE
- Calculate R² Score
- Calculate Adjusted R²
- Write the Linear Regression equation
- Interpret the model coefficients

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# 1. Train the Linear Regression Model
model = LinearRegression()
model.fit(X_train, y_train)

# 2. Make Predictions on the Test Set
y_pred = model.predict(X_test)

# 3. Calculate Performance Metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Calculate Adjusted R2
n = X_test.shape[0]  # Number of observations in test set
p = X_test.shape[1]  # Number of features
adjusted_r2 = 1 - ((1 - r2) * (n - 1) / (n - p - 1))

# --- Display Evaluation Metrics ---
print("--- MODEL EVALUATION METRICS ---")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R² Score: {r2:.4f}")
print(f"Adjusted R² Score: {adjusted_r2:.4f}\n")

# --- Display Intercept and Coefficients ---
print("--- MODEL PARAMETERS ---")
print(f"Intercept (Beta_0): {model.intercept_:.4f}\n")

print("Coefficients for individual features:")
for feature, coef in zip(X.columns, model.coef_):
    print(f" * {feature}: {coef:.4f}")
```

$Interpreting\ the\ Model\ Coefficients$
